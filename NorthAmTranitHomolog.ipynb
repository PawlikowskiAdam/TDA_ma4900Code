```python
import numpy as np
import gudhi as gd
import pandas as pd
from sklearn import manifold
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.tri as mtri
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

Data and vietoris rips complex with graphs

```python
stations_CTA = np.loadtxt('data/points/CTA_L_X_Y.txt', delimiter=",", skiprows=1)
stations_MTA = np.loadtxt('data/points/MTA_X_Y.txt', delimiter=",")
stations_TTC = np.loadtxt('data/points/TTC_C_Y.txt', delimiter=",")
stations_WMATA = np.loadtxt('data/points/WMATA_X_Y.txt', delimiter=",")

fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 20))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# ---------------- CTA ----------------
rips_CTA = gd.RipsComplex(points=stations_CTA)
tree_CTA = rips_CTA.create_simplex_tree(max_dimension=2)
barcodes_CTA = tree_CTA.persistence()

axes[0,0].scatter(stations_CTA[:,0], stations_CTA[:,1], s=10)
axes[0,0].set_title("CTA Stations")

gd.plot_persistence_diagram(barcodes_CTA, axes=axes[0,1])
axes[0,1].set_title("CTA Persistence Diagram")

gd.plot_persistence_density(barcodes_CTA, axes=axes[0,2])
axes[0,2].set_title("CTA Persistence Density")

# ---------------- MTA ----------------
rips_MTA = gd.RipsComplex(points=stations_MTA)
tree_MTA = rips_MTA.create_simplex_tree(max_dimension=2)
barcodes_MTA = tree_MTA.persistence()

axes[1,0].scatter(stations_MTA[:,0], stations_MTA[:,1], s=10)
axes[1,0].set_title("MTA Stations")

gd.plot_persistence_diagram(barcodes_MTA, axes=axes[1,1])
axes[1,1].set_title("MTA Persistence Diagram")

gd.plot_persistence_density(barcodes_MTA, axes=axes[1,2])
axes[1,2].set_title("MTA Persistence Density")

# ---------------- TTC ----------------
rips_TTC = gd.RipsComplex(points=stations_TTC)
tree_TTC = rips_TTC.create_simplex_tree(max_dimension=2)
barcodes_TTC = tree_TTC.persistence()

axes[2,0].scatter(stations_TTC[:,0], stations_TTC[:,1], s=10)
axes[2,0].set_title("TTC Stations")

gd.plot_persistence_diagram(barcodes_TTC, axes=axes[2,1])
axes[2,1].set_title("TTC Persistence Diagram")

gd.plot_persistence_density(barcodes_TTC, axes=axes[2,2])
axes[2,2].set_title("TTC Persistence Density")

# ---------------- WMATA ----------------
rips_WMATA = gd.RipsComplex(points=stations_WMATA)
tree_WMATA = rips_WMATA.create_simplex_tree(max_dimension=2)
barcodes_WMATA = tree_WMATA.persistence()

axes[3,0].scatter(stations_WMATA[:,0], stations_WMATA[:,1], s=10)
axes[3,0].set_title("WMATA Stations")

gd.plot_persistence_diagram(barcodes_WMATA, axes=axes[3,1])
axes[3,1].set_title("WMATA Persistence Diagram")

gd.plot_persistence_density(barcodes_WMATA, axes=axes[3,2])
axes[3,2].set_title("WMATA Persistence Density")

plt.show()
```

```python
# ---------------- CTA ----------------
Ista1_CTA = tree_CTA.persistence_intervals_in_dimension(1)
Ista0_CTA = tree_CTA.persistence_intervals_in_dimension(0)
lifetimes1_CTA = Ista1_CTA[:,1] - Ista1_CTA[:,0]
order_CTA = np.argsort(lifetimes1_CTA)[::-1]
order0_CTA = np.sort(Ista0_CTA)[::-1]
Ista_sorted1_CTA = Ista1_CTA[order_CTA]
#print(Ista_sorted1_CTA)
#print(order0_CTA)

# ---------------- MTA ----------------
Ista1_MTA = tree_MTA.persistence_intervals_in_dimension(1)
Ista0_MTA = tree_MTA.persistence_intervals_in_dimension(0)
lifetimes1_MTA = Ista1_MTA[:,1] - Ista1_MTA[:,0]
order_MTA = np.argsort(lifetimes1_MTA)[::-1]
order0_MTA = np.sort(Ista0_MTA)[::-1]
Ista_sorted1_MTA = Ista1_MTA[order_MTA]
#print(Ista_sorted1_MTA)
#print(order0_MTA)

# ---------------- TTC ----------------
Ista1_TTC = tree_TTC.persistence_intervals_in_dimension(1)
Ista0_TTC = tree_TTC.persistence_intervals_in_dimension(0)
lifetimes1_TTC = Ista1_TTC[:,1] - Ista1_TTC[:,0]
order_TTC = np.argsort(lifetimes1_TTC)[::-1]
order0_TTC = np.sort(Ista0_TTC)[::-1]
Ista_sorted1_TTC = Ista1_TTC[order_TTC]
#print(Ista_sorted1_TTC)
#print(order0_TTC)

# ---------------- WMATA ----------------
Ista1_WMATA = tree_WMATA.persistence_intervals_in_dimension(1)
Ista0_WMATA = tree_WMATA.persistence_intervals_in_dimension(0)
lifetimes1_WMATA = Ista1_WMATA[:,1] - Ista1_WMATA[:,0]
order_WMATA = np.argsort(lifetimes1_WMATA)[::-1]
order0_WMATA = np.sort(Ista0_WMATA)[::-1]
Ista_sorted1_WMATA = Ista1_WMATA[order_WMATA]
#print(Ista_sorted1_WMATA)
print(order0_WMATA)
```

```python
systems = [
    ("CTA", stations_CTA, tree_CTA, [0.057265, 0.057266, 0.049405, 0.049406]),
    ("MTA", stations_MTA, tree_MTA, [0.054041, 0.054042, 0.040797, 0.040798]),
    ("TTC", stations_TTC, tree_TTC, [0.026306, 0.0263069, 0.024667, 0.02466799]),
    ("WMATA", stations_WMATA, tree_WMATA, [0.186428,0.1864283,0.176952, 0.17695288])
]

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(36, 30))
plt.subplots_adjust(hspace=0.4, wspace=0.3)
for row, (name, stations, sta_tree, eps_list) in enumerate(systems):
    for col, epsilon in enumerate(eps_list):
        ax = axes[row, col]
        edges = []
        for simplex, filt in sta_tree.get_skeleton(2):
            if filt <= epsilon and len(simplex) == 2:
                edges.append(simplex)

        edges = np.array(edges)
        for e in edges:
            x = stations[e,0]
            y = stations[e,1]
            ax.plot(x, y, color='blue', lw=0.6)
        ax.scatter(stations[:,0], stations[:,1], color='red', s=5)
        ax.set_aspect('equal')
        ax.set_title(f"{name} ε={epsilon}")

plt.tight_layout()
plt.show()
```
